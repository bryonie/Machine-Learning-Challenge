```python
import pandas as pd
import numpy as np
import librosa
```

```python
path = np.load("path.npy")
feat = np.load("feat.npy")
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
#creating dataframe feat
df_feat = pd.DataFrame(feat)
pd.set_option('display.max_colwidth', -1)
df_feat.head(1)
#pd.reset_option("^display") 
```

```python
#checking distribution labels
train['word'].value_counts()
```

```python
#number of frames for each signal #indifferent
def number_frames(argument):
    for x in argument:
        count=0
        for y in x:
            count +=1
        print(count)

number_frames(feat)
```

```python
#check if every frame has 13 coefficients
def coefficients(argument):
    for x in argument:
        for y in x:
            if len(y) == 13:
                continue
            else:
                return False

coefficients(feat)
```

```python
#computing delta coefficienten
delta_values = list()
for x in feat:
    delta_coef = librosa.feature.delta(x)
    delta_values.append(delta_coef)
delta_array= np.array(delta_values)

delta_array
```

```python
type(delta_array)
```

```python
#computing delta delta coefficienten
delta_delta_values = list()
for x in feat:
    delta_delta_coef = librosa.feature.delta(x, order=2)
    delta_delta_values.append(delta_delta_coef)
delta_delta_array= np.array(delta_delta_values)

delta_delta_array
```

```python
#creating dataframe from path array
df_path = pd.DataFrame(path)
#adding column name
df_path.columns = ["path"]
df_path
```

```python
#joining df_feat with df_path
feat_and_path = pd.concat([df_feat,df_path], axis=1, join="inner")
#defining name columns
feat_and_path.columns = ["coefficients", "path"]
feat_and_path
```

```python
#creating training dataframe with coefficients and label
train_features_label = train.merge(feat_and_path, how = "inner", on = ["path"])
train_features_label = train_features_label.drop(columns = "path")
train_features_label
```

```python
#creating test dataframe with coefficients
test_features= test.merge(feat_and_path, how = "inner", on = ["path"])
test_features = test_features.drop(columns = "path")
test_features
```
